## Multiple Linear Regression

Now you know how to build a model with one X (feature variable) and Y (response variable). But what if you have three feature variables, or may be 10 or 100? Building a separate model for each of them, combining them, and then understanding them will be a very difficult and next to impossible task. By using multiple linear regression, you can build models between a response variable and many feature variables.

Let's see how to do that.

### Step_3: Splitting the Data for Training and Testing

```python
# Putting feature variable to X
X = advertising_multi[['TV','Radio','Newspaper']]

# Putting response variable to y
y = advertising_multi['Sales']
```

```python
#random_state is the seed used by the random number generator. It can be any integer.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7 , random_state=100)
```

### Step_4 : Performing Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
# Representing LinearRegression as lr(Creating LinearRegression Object)
lm = LinearRegression()
```

```python
# fit the model to the training data
lm.fit(X_train,y_train)
```

### Step_5 : Model Evaluation

```python
# print the intercept
print(lm.intercept_)
```

```python
# Let's see the coefficient
coeff_df = pd.DataFrame(lm.coef_,X_test.columns,columns=['Coefficient'])
coeff_df
```

From the above result we may infern that if TV price increses by 1 unit it will affect sales by 0.045 units.

### Step_6 : Predictions

```python
# Making predictions using the model
y_pred = lm.predict(X_test)
```

### Step_7: Calculating Error Terms

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r_squared = r2_score(y_test, y_pred)
```

```python
print('Mean_Squared_Error :' ,mse)
print('r_square_value :',r_squared)
```

### Checking for P-value Using STATSMODELS

> p-value is used to vaalidate if the features we are using in our model are statistically significant or not.

```python
import statsmodels.api as sm
X_train_sm = X_train
#Unlike SKLearn, statsmodels don't automatically fit a constant, 
#so you need to use the method sm.add_constant(X) in order to add a constant. 
X_train_sm = sm.add_constant(X_train_sm)
# create a fitted model in one line
lm_1 = sm.OLS(y_train,X_train_sm).fit()

# print the coefficients
lm_1.params
```

```python
print(lm_1.summary())
```

>We can see that Newspaper has high p-value. High p-value suggests that, our null hypothesis is true. This imples Beta for this feature is zero athis means, newspaper is insignificant.

### Simple Linear Regression: Newspaper(X) and Sales(y)

```python
import pandas as pd
import numpy as np
# Importing dataset
advertising_multi = pd.read_csv('advertising.csv')

x_news = advertising_multi['Newspaper']

y_news = advertising_multi['Sales']

# Data Split
from sklearn.model_selection import train_test_split
X_train_temp, X_test_temp, y_train_temp, y_test_temp = train_test_split(x_news, y_news, 
                                                    train_size=0.7 , 
                                                    random_state=110)

# Required only in the case of simple linear regression
X_train_temp = X_train_temp[:,np.newaxis]
X_test_temp = X_test_temp[:,np.newaxis]

# Linear regression from sklearn
from sklearn.linear_model import LinearRegression
lm = LinearRegression()

# Fitting the model
lm.fit(X_train_temp,y_train_temp)

# print params 
print(lm.intercept_)
print(lm.coef_)

# Making predictions
y_pred_temp = lm.predict(X_test_temp)

# Importing mean square error and r square from sklearn library.
from sklearn.metrics import mean_squared_error, r2_score

# Computing mean square error and R square value
mse = mean_squared_error(y_test_temp, y_pred_temp)
r_squared = r2_score(y_test_temp, y_pred_temp)

# Printing mean square error and R square value
print('Mean_Squared_Error :' ,mse)
print('r_square_value :',r_squared)
```

> We can see that Newspaper has the coefficient same as what we got for the TV earlier. Also, with the R_square_value we can see that it can explain variance of the 8% data.

### Let's now understand about the model evaluation metric adjusted r-squared and why that is a more trustworthy metric than r-squared. 

Adjusted R-sqared tries to penalize model when we have too many variables. So when we have too many features, adjusted r-sqared value penalizes model and its value is reduced. 

In the previous few sections we saw that we might use adjusted r-square and P-value to choose the best set of predictors for our model. We can also use a simple method for selecting variables, i.e. RFE(Recursive Feature Elimination). The goal of recursive feature elimination (RFE) is to choose features by recursively considering smaller and smaller sets of features.

![alt text](adjusted-r2.png "Title")


The adjusted R-square make sure that the new variable added should be significant to explain the dependent variable and not just by chance. 

> To decide the optimal number of features, run RFE multiple times with different number of features and plot the test-adjusted-r2 value and decide based on that.

### Model Refinement Using RFE

The goal of recursive feature elimination (RFE) is to select features by recursively considering smaller and smaller sets of features. First, the estimator is trained on the initial set of features and the importance of each feature is obtained either through a coef_ attribute or through a feature_importances_ attribute. Then, the less important features are pruned from the the current set of features. This procedure is recursively repeated on the pruned dataset until the desired number of features to select is reached.

```python
from sklearn.feature_selection import RFE
```

```python
rfe = RFE(lm, step=2)
```

```python
rfe = rfe.fit(X_train, y_train)
```

```python
print(rfe.support_)
print(rfe.ranking_)
```
